```python
import numpy as np
import pandas as pd
import random
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Conv2D, Dense, MaxPooling2D
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist

# Load data
(X_train, y_train), (X_test, y_test) = mnist.load_data()
print(X_train.shape)  # (60000, 28, 28)
print(X_train[0].min(), X_train[0].max())  # np.uint8 0 ~ 255

# Normalize data
X_train = X_train / 255.0
X_test = X_test / 255.0
print(X_train[0].min(), X_train[0].max())  # np.float64 0.0 ~ 1.0

# (Optional) Set figure size for plotting
plt.figure(figsize=(16, 10))

# Function to plot a single digit
def plot_digit(image, digit, i):
    plt.subplot(4, 5, i + 1)
    plt.imshow(image, cmap='gray')
    plt.title(f"Digit {digit}")
    plt.xticks([])
    plt.yticks([])

# Plot the first 20 images
for i in range(20):
    plot_digit(X_train[i], y_train[i], i)
plt.show()

# Reshape for CNN input
X_train = X_train.reshape(X_train.shape + (1,))
X_test = X_test.reshape(X_test.shape + (1,))

# Print first 20 labels
print("First 20 labels of y_train:", y_train[:20])

# Build the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(100, activation='relu'),
    Dense(10, activation='softmax')
])

optimizer = SGD(learning_rate=0.01, momentum=0.9)
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32)

# Evaluate the model
score = model.evaluate(X_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])

# Visualize model performance (accuracy & loss)
fig = plt.figure(figsize=(10, 8))
# Plot accuracy
plt.subplot(2, 1, 1)
plt.plot(model.history.history['accuracy'], label='Train')
plt.plot(model.history.history.get('val_accuracy', []), label='Validation')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='lower right')
# Plot loss
plt.subplot(2, 1, 2)
plt.plot(model.history.history['loss'], label='Train')
plt.plot(model.history.history.get('val_loss', []), label='Validation')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

# Plot 20 random test images with predictions
def plot_digit_predict(image, digit, i):
    plt.subplot(4, 5, i + 1)
    plt.imshow(image, cmap='gray')
    plt.title(f"Predicted {digit}")
    plt.xticks([])
    plt.yticks([])

for i in range(20):
    image = random.choice(X_test).squeeze()
    pred = model.predict(image.reshape(1, 28, 28, 1))
    digit = np.argmax(pred[0])
    plot_digit_predict(image, digit, i)
plt.show()

# Save model architecture and weights
model_digit_json = model.to_json()
with open("model_digit.json", "w") as json_file:
    json_file.write(model_digit_json)
model.save_weights("model_digit.weights.h5")
print("Saved model to disk")
```
